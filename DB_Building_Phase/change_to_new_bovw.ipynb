```python
import json
import pandas as pd
```

```python
new_bovw = json.load(open("../data_files/bovw_pack(1).json"))
img_ids = pd.read_csv("../data_files/export.csv")["image_Id"].tolist()
```

```python
print(new_bovw[0])
```

```python
print(len(new_bovw), len(img_ids))
```

```python
copy_bovw = new_bovw.copy()
def check_id_in_bovw(image_id):
    for i in range(len(copy_bovw)):
        if copy_bovw[i]["ID"] == image_id:
            copy_bovw.pop(i)
            return True
    return False
```

```python
#check if the image ids are in the new bovw if they are not in the new bovw, remove them from the bovw
i = 0
for img_id in img_ids:
    if check_id_in_bovw(img_id):
        print(i, "found")
        i += 1
    else:
        print(f"image id {img_id} not found in bovw")
```

```python
print("new bovw length", len(copy_bovw))
#make a new bovw that removes the image ids that are not in img_ids
new_bovw = new_bovw.copy()
for object in copy_bovw:
    new_bovw.remove(object)
print("new bovw length", len(new_bovw))
```

```python
#save the new bovw to a json file
with open("../data_files/new_bovw.json", "w") as f:
    json.dump(new_bovw, f, indent=4)
```
